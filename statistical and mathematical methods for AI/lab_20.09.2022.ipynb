```python
import numpy as np
import matplotlib.pyplot as plt

# Creating two vectors
a = 0
b = 2*np.pi
N = 50

x = np.linspace(a, b, N)
y1 = np.sin(x)
y2 = np.tan(x)


# Visualize
plt.plot(x, y1, 'd', color = "red", markersize = 2)
plt.plot(x, y2)
plt.title('A plot of trig. functions.')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([-4, 4])
plt.legend(['f(x) = sin(x)', 'f(x) = tan(x)'])
plt.grid()
plt.show()
```

```python
import numpy as np
import scipy
A = np.array([[1, 2, 4],[2, 1, 3], [1, 0, 0]])
y = np.array([0,1,0])
# Define a function that solves the system
def solve(A, y):
    # LU factorization of A
    P, L, U = scipy.linalg.lu(A)

    # Solve Lz = Py
    z = scipy.linalg.solve_triangular(L, P@y, lower = True)

    # Solve Ux = z
    x = scipy.linalg.solve_triangular(U, z)

    return x

print (solve(A, y) ) 
```
